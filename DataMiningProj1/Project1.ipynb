```python
shooter_df = game_df[game_df['Genre'] == 'Shooter']
shooter_totalsales = shooter_df[['NA_Sales','EU_Sales','JP_Sales']].sum()
```

```python
sales_labels = ['North America Sales', 'Europe Sales', 'Japan Sales']
sales_values = shooter_totalsales.values

plt.figure(figsize=(10, 6))
plt.bar(sales_labels, sales_values, color=['blue', 'green', 'red'])
plt.xlabel('Sales Regions')
plt.ylabel('Total Sales')
plt.title('Total Sales for Shooter Genre by Region')

def millions_formatter(x, pos):
    """Formatter function to convert y-axis values to millions with 'm' suffix."""
    return f'{int(x)}m'

plt.gca().yaxis.set_major_formatter(FuncFormatter(millions_formatter))  

plt.show()
```

```python
game_df
```

```python
agg_function = {
    'Rank': 'first',    
    'Platform': 'first',         
    'Year': 'min',     
    'Genre': 'first',           
    'Publisher': 'first',      
    'NA_Sales': 'sum',           
    'EU_Sales': 'sum',          
    'JP_Sales': 'sum',           
    'Other_Sales': 'sum',        
    'Global_Sales': 'sum'        
}

combined_df = game_df.groupby('Name').aggregate(agg_function).reset_index()
```

```python
combined_df
```

```python
top_sales_idxmax = combined_df.groupby('Year')['Global_Sales'].idxmax()
```

```python
top_sales_df = combined_df.loc[top_sales_idxmax].reset_index(drop=True)
```

```python
pd.set_option('display.max_columns', None)  
pd.set_option('display.max_rows', None)     
pd.set_option('display.max_colwidth', 20)   
pd.set_option('display.width', 1000) 
```

```python
top_sales_df
```

```python
top_sales_df= top_sales_df.drop([37,38])
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(x='Year', y='Global_Sales', data=top_sales_df)
plt.title('Most Games Sold Each Year')
plt.xlabel('Year')
plt.ylabel('Global Sales')
plt.show
```

```python
annual_sales = combined_df.groupby('Year')['Global_Sales'].sum().reset_index()
annual_sales.drop([37,38])
```

```python
plt.figure(figsize=(16,6))
sns.lineplot(x='Year', y='Global_Sales', data=annual_sales, marker='o', color='b', linestyle='-', linewidth=2)

plt.title('Total Global Sales Over Time')
plt.xlabel('Year')
plt.ylabel('Total Global Sales (in millions)')
plt.grid(True)
plt.xticks(annual_sales['Year']) 
plt.tight_layout()
plt.xticks(rotation=45) 

plt.show()
```
